```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt

ob=pd.read_csv("C:\\Users\\sibag\\OneDrive\\Desktop\\my file\\EDA project\\google_cleaned (2).csv")
df=pd.DataFrame(ob)

# Top 5 most installs Category
installs=df.groupby('Category')['Installs'].sum().sort_values(ascending=False)
top_installs=installs.head(5)

mlt.figure()
mlt.bar(top_installs.index,top_installs.values)
mlt.xticks(rotation=90)
mlt.xlabel("Category")
mlt.ylabel("Total Installs")
mlt.title("Top 5 Categories  by Total Installs")
mlt.show()

```
